```python
import os
os.chdir('/home/angel/Downloads/ProjetI&A')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('list_attr_celeba.csv')
```

```python
#First we check if we have some missing values
df.isnull().sum(axis = 0)
```

```python
#Once we select the feauture to train our neural network with, we filter our dataset
df = df[['image_id','Eyeglasses']]
```

```python
df
```

```python
df_oui = df.query('Eyeglasses==1')
```

```python
df_x = df.iloc[:,:-1]
```

```python
df_y = df.iloc[:,-1]
```

```python
X = df_x.values
Y = df_y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
indexes_oui_test = [i for i,x in enumerate(Y_test) if x == 1]
indexes_non_test = [i for i,x in enumerate(Y_test) if x == -1]
```

```python
indexes_oui_train = [i for i,x in enumerate(Y_train) if x == 1]
indexes_non_train = [i for i,x in enumerate(Y_train) if x == -1]
```

```python
x_test_oui,x_test_non,x_train_oui,x_train_non = [],[],[],[]
for i in indexes_oui_test:
    x_test_oui.append(X_test[i])
for i in indexes_non_test:
    x_test_non.append(X_test[i])
for i in indexes_oui_train:
    x_train_oui.append(X_train[i])
for i in indexes_non_train:
    x_train_non.append(X_train[i])
```

```python
import shutil
c = np.concatenate(x_train_oui,axis = 0)
for x in c:
    shutil.copy('img_align_celeba/'+x,'Train/Presence_of_feature')
```

```python
x_train_non = np.concatenate(x_train_non,axis = 0)
x_test_non = np.concatenate(x_test_non,axis = 0)
x_test_oui = np.concatenate(x_test_oui,axis = 0)
```

```python
for x in x_train_non:
    shutil.copy('img_align_celeba/'+x,'Train/Absence_of_feature')
for x in x_test_oui:
    shutil.copy('img_align_celeba/'+x,'Test/Presence_of_feature')
for x in x_test_non:
    shutil.copy('img_align_celeba/'+x,'Test/Absence_of_feature')
```

```python
pd.DataFrame(x_test_non).to_csv("x_test_non.csv")
```

```python
pd.DataFrame(x_test_oui).to_csv("x_test_oui.csv")
pd.DataFrame(x_train_non).to_csv("x_train_non.csv")
pd.DataFrame(x_train_oui).to_csv("x_train_oui.csv")
```
